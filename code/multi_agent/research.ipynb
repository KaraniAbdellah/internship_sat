```python
# Import Packages
from langchain.agents import create_agent
from langchain.tools import tool
from dotenv import load_dotenv
import os
from tavily import TavilyClient
from pprint import pprint
```

```python
load_dotenv(".env")
TAVILY_API_KEY = os.getenv("TAVILY_API_KEY")
GROQ_API_KEY = os.getenv("GROQ_API_KEY")
```

```python
tavily_client = TavilyClient(api_key=TAVILY_API_KEY)
```

```python
# Define Tools
@tool
def search_web(query:str, num_results:int = 3):
    '''
        Search the web and get back a list search results including the page, title, url and a short summary of each webpage.
        
        Args:
            query: The search query
            num_results: The number of results to return, max is 3
        
        Returns:
            A dictionary of search results
    '''
    response = tavily_client.search(query=query, num_results=num_results)
    processed_result = {
        "query": query,
        "results": []
    }
    for res in response["results"]:
        processed_result["results"].append(res["url"])
        processed_result["results"].append(res["title"])
        processed_result["results"].append(res["content"])
    return processed_result
```

```python
res = search_web.func("What best players in world ?")
pprint(res)
```
